```
3. 1~100까지 평균 구하기
```



````python
i = 0
for ii in range(1, 101):
  i += ii
print(i/ii)
````

````python
list3 = list(range(1, 101))
print(list3)

list3_2 = []
for i in range(1, 101):
  list3_2.append(i)
print(list3_2)

sum_list3 = sum(list3)
print(sum_list3)

len_list3 = len(list3)
print(len(list3))

avg = sum_list3/len_list3
print(avg)
````

```
4. 구구단 출력하기
```



````python
for i in range(1, 10):
  print(9,"x",i,"=",i*9)
````

````python
for i in range(1, 10):
  for j in range(1, 10):
    print(i,"x",j,"=",i*j)
  print("\n","----------")
````

````python
for i in range(1, 10):
  for j in range(1, 10):
    print(f"{i}*{j}=",i*j)
````

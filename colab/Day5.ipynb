```python
import os
import requests
import time
import threading

folder = 'images'
os.makedirs(folder, exist_ok = True)

def img_download(url, folder):
  file_name = url.split('/')[-1] +'.jpg'
  file = os.path.join(folder, file_name)
  img = requests.get(url).content
  with open(file, 'wb') as f:
    f.write(img)

for url in my_urls:
 thread = threading.Thread(target=img_download, args=(url, folder))
 thread.start()

thread.join()
```

```python
import cv2
from google.colab.patches import cv2_imshow


file = '/content/cat2.jpg'
img = cv2.imread(file)
cv2_imshow(img)
print(img)

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
print(img_gray)
```

```python
print(img.shape)

h = img.shape[0]
print(h)

aspect_r = 350 /h
print(aspect_r)

w = img.shape[1]
print(w)

img_l = cv2.resize(img, ( int(w * aspect_r), 350))
cv2_imshow(img_l)
```

```python
# cv2 이미지는 bgr 형태
# bgr -> rgb

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2_imshow(img_rgb)
```

```python
img_copy = img.copy()
#cv2_imshow(img_copy)

h = img.shape[0]
w = img.shape[1]
cv2.line(img_copy, (0,0), (w, h), (0, 255, 0), 3, cv2.LINE_AA)
cv2_imshow(img_copy)
```

```python
img_copy2 = img.copy()
cv2.circle(img_copy2,
        (int(w/2), int(h/2)),
        20, (0, 0, 255), -1)   # -1대신 0이면 빈 원
cv2_imshow(img_copy2)
```

```python
img_copy3 = img.copy()
cv2.rectangle(img_copy3, (int(w/5), int(h/5)), (int(w/5)*4 , int(h/5)*4),
 (255, 255, 0), 5)
cv2_imshow(img_copy3)
```

```python
 # global 전역변수

hp = 100
def attack(num):
   global hp
   hp -= num

def recovery(num):
  global hp
  hp += num

def reset():
  global hp
  hp = 100

## 총알의 위치가 캐릭터에 닿으면
attack(30)
print(hp)
recovery(10)
print(hp)
reset()
print(hp)
```

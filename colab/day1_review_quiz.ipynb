````python
students = ["Alice", "Bob", "Charlie"]
scores = [90, 85, 80]

for i in range(3):
  print(f'{students[i]}의 점수 : {scores[i]}')

print(f"평균 점수 : {sum(scores)/len(scores)}")
````

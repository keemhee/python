1~100까지 정수 중 짝수만 출력해주는 함수 print_even, 홀수만 출력해주는 함수 print_odd를 생성하여 2개의 Thread를 생성하여 동시에 실행하도록 하고 Thread의 작업이 끝날때 까지 대기 하였다가 "Thread 작업 종료" 메시지 출력하세요.

```python
import threading

#짝수 출력 함수
def print_even():
  for i in range(1, 101):
    if i % 2 == 0:
      print(i)

#홀수 출력 함수
def print_odd():
  for i in range(1, 101):
    if i % 2 == 1:
      print(i)

#스레드 생성
thread1 = threading.Thread(target=print_even)
thread2 = threading.Thread(target=print_odd)

#스레드 시작
thread1.start()
thread2.start()

#스레드 종료까지 대기
thread1.join()
thread2.join()

print("Thread 작업종료")
```

1~10까지의 int형 데이터를 랜덤으로 생성하여 리스트에 저장하고, 저장된 리스트에서 각 수의 빈도수를 구하고 matplotlib를 이용하여 막대그래프로 시각화하세요

```python
import random
import matplotlib.pyplot as plt

#랜덤 리스트 생성
random_list = [random.randint(1, 10) for _ in range(10)]

#빈도수 dic 생성
num_dic = {}
for i in random_list:
  if i not in num_dic.keys():
    num_dic[i] = 1
  else:
    num_dic[i] += 1

#plt bar의 키와 값을 bar의 데이터로 변경
keys = list(num_dic.keys())
values = list(num_dic.values())

plt.bar(keys, values)
```

(300, 300,3) 크기의 numpy배열 생성 후 start_point = (25,25) end_point(250,250)크기의 사각형을 그리고 TrackBar를 이용하여 B,G,R값을 실시간으로 변경 하는 코드를 작성하세요 (q 버튼을 누르면 종료됩니다)

```python
import cv2
import numpy as np

#전체 배열 생성
rectangle = np.zeros((300, 300, 3), dtype='uint8')

def do_nothing(value):
    pass

#트랙바 생성
cv2.namedWindow("Track_bar")
cv2.createTrackbar("color_B", "Track_bar", 0, 255, do_nothing)
cv2.createTrackbar("color_G", "Track_bar", 0, 255, do_nothing)
cv2.createTrackbar("color_R", "Track_bar", 0, 255, do_nothing)

#트랙바에서 실시간으로 B,G,R값 받아 색상 변경
while True:
    B_value = cv2.getTrackbarPos("color_R", "Track_bar")
    G_value = cv2.getTrackbarPos("color_G", "Track_bar")
    R_value = cv2.getTrackbarPos("color_B", "Track_bar")

    bgr = (B_value, G_value, R_value)
    cv2.rectangle(rectangle, (25,25,250,250), bgr, -1, 0)

    cv2.imshow("Rect", rectangle)

#q 누르면 종료
    key = cv2.waitKey(5) & 0xFF
    if key == ord('q'):
        break

cv2.destroyAllWindows()
```

```python
def report():
  my_students = ["흥민", "석구", "우성", "명신"]
  grades = [80, 75, 93, 15]

  grades1 = {}

  for s, g in zip(my_students, grades):
    if g >= 90:
      grades1[s] = "A"
    elif g >= 80:
      grades1[s] = "B"
    elif g >= 70:
      grades1[s] = "C"
    else:
      grades1[s] = "F"

  for key in grades1:
    print(f"{key} : {grades1[key]}")

report()
```

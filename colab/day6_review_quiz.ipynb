```python
#규정속도 speed
#도로를 달리는 모든 차의 속도가 담긴 리스트 cars


cars = [110, 98, 125, 148, 120, 112, 89]

def solution(speed ,cars):
  i_cars = []
  for i in cars:
    if i > speed:
      i_cars.append(i)
  fine = 0
  for s in i_cars:
    if 1.1*speed <= s < 1.2*speed:
      fine += 3
    elif 1.2*speed <= s < 1.3*speed:
      fine += 5
    elif 1.3*speed <= s < 200:
      fine += 7
    else:
      fine = fine
  return fine
#print(len(i_cars))
solution(100, cars)
```

```python
# Class는 빵틀
class MyClass:
  pass
```

```python
# 그냥 클래스 선언
myClass1 = MyClass()
myClass2 = MyClass()
```

```python
# 속성 만들기 이해하기

myClass1.first_name = "Halona"
myClass2.first_name = "Keanue"
myClass1.last_name = "Yun"
myClass2.last_name = "Kim"
```

```python
# 이 클래스 안에서 갖춰야하는 것
# init의 first_name, last_name이 파라미터1, 파라미터2, 콤마찍고 뒤에 더 쓰면 파라미터 다수 가능
class MyClass:
  def __init__(self, first_name, last_name):   #self는 클래스의 속성이니 그냥 꼭 하나 넣어줘야 하는 거
    self.first_name = first_name               #first_name, last_name을 받아야 생성이 될 수 있게
    self.last_name = last_name

  def method1(self):
    print("Method1 입니다")

  def printFullname(self):
    print(f"{self.last_name} {self.first_name}입니다") #class 안에서는 클래스 이름을 self라고 부름
```

```python
myClass1 = MyClass("Halona", "Yun")         #재료를 주고 들어가는
myClass2 = MyClass("HeeJi", "Kim")          #객체를 만듦
```

```python
#이제 속성값으로 지정해줘서 나옴
print(myClass1.first_name)
print(myClass2.last_name)

myClass1.method1()
myClass2.method1()
```

```python
#성과 이름 한 번에 프린트
print(f"{myClass1.last_name} {myClass1.first_name}입니다")
myClass1.printFullname()
myClass2.printFullname()

#빵틀을 끌어내고 할 수 있는 걸 블록처럼 만들어놓는다
```

```python
#calculator 만들기

class Calculator():
  def __init__(self):
    self.result = 0     #받아서 받을 수도 있고, 지정할 수도 있음

  def add(self, num1, num2):    #두개의 수만 받아서 계산한다
    self.result = num1 + num2   #return 값이 필요없음

  def multiply(self, num1, num2, num3):    #이 함수 안에서만 해당하는 거라 위에 거랑 이름 같아도 괜찮음
    self.result = num1 * num2 * num3
```

```python
cal1 = Calculator()
cal1.result
# 객체 생성 확인

cal1.add(3, 4)
print(cal1.result)
cal1.multiply(1, 2, 3)
print(cal1.result)
```

```python
#신호등 클래스

colors = ['green', 'yellow', 'red']

class TrafficLight:
  def __init__(self, color):

    self.color = color

  def action(self):
    if self.color == 'red':
      self.next_color = 'green'
      print(f'{self.color}입니다. 멈추세요.')
     #print(f"다음 신호는 {next_color}입니다.")
    elif self.color == 'yellow':
      self.next_color = 'red'
      print(f'{self.color}입니다. 멈춤 준비하세요.')
      #print(f"다음 신호는 {next_color}입니다.")
    elif self.color == 'green':
      self.next_color = 'yellow'
      print(f'{self.color}입니다. 가세요.')
      #print(f"다음 신호는 {next_color}입니다.")

      #self없으면 그 함수 method안에서만 인지.
```

```python
for c in colors:
  tl = TrafficLight(c)
  #print(tl.color)
  tl.action()
  print(f"다음 신호는 {tl.next_color}입니다.")
  #위에 action의 next_color에 self 붙여줘야 여기서 next_color 부를 수 있음.
  print('\n')
```

```python
class Employee:

  #variable 고정하면
  pay_ratio = 1.3

  def __init__(self, first_name, last_name, pay):
    self.first_name = first_name
    self.last_name = last_name
    self.pay = pay
    self.email = self.first_name.lower() + '_' + self.last_name.lower() + '@openAI.com'

  def raise_pay(self):
    self.pay = self.pay * self.pay_ratio

  def print_fullname(self):
    print(f"{self.last_name} {self.first_name}")
```

```python
#emp1.email 치면 이메일이 나오게끔
#이메일은 firstname_lastname@openAI.com
emp1 = Employee('HeungMin', 'Son', 45000)
print(emp1.email)

#1년후 연봉
print(emp1.pay)
emp1.raise_pay()
print(emp1.pay)

emp1.raise_pay()
print(emp1.pay)
```

```python
#상속
class  Developer(Employee): #전체 employee를 상속받음. 속성과 메서드를 받을 수 있는 퍼미션.
  def __init__(self, first_name, last_name, pay, coding):
    super().__init__(first_name, last_name, pay)  #위 양식대로 하면 위에서 했던 거 데려옴. super는 부모 클래스를 불러오는 거.  여기선 Employee거 다 사용가능.
    self.coding = coding

dev1 = Developer("Keanue","Kim",50000,"Python")

print(dev1.pay)
print(dev1.coding)
print(dev1.email)

dev1.raise_pay()
print(dev1.pay)
```

```python
#직원 리스트화
#manager의 권한을 받은 사람만 add_employee할 수 있게
#매니저가 처음 만들면서 누군가를 받으면 그 사람을 집어넣고
#아무도 안 받으면 그냥 매니저 역할만 시작하는.

class Manager(Employee):

  def __init__(self, first_name, last_name, pay, employees=None):    #none은 꼭 전달을 하지 않아도 되는 아이
    super().__init__(first_name, last_name, pay)
    if employees == None:
      self.employees =[]
      #임플로이를 받을 수도 있고 안 받을 수도 있고
    else:
      self.employees = employees
    print(self.employees)

  def add_employee(self, emp):
    if emp not in self.employees:
      self.employees.append(emp)
      #매니저권한으로 근로자 추가

  def print_employees(self):
    for emp in self.employees:
      #프린트 이름
      emp.print_fullname()


manager = Manager("Halona", "Yun", 100000,[dev1]) #할로나를 매니저로 만듦.   #첫번째는 리스트로 전해줘야함.
print(manager.first_name)                  #매니저의 이름은 할로나다.
```

```python
manager.add_employee(emp1)
manager.print_employees()
```

```python
class Person:
  def __init__(self, firstname, lastname, gender):
    self.firstname = firstname
    self.lastname = lastname
    self.gender = gender
    self.email = self.firstname.lower()+self.lastname.lower()+"@aaa.com"
    self.name = self.lastname + self.firstname
  #이름  성별  이메일  print()
  def info(self):
    print(self.name, self.gender, self.email)
```

```python
p1 = Person("HeungMin", "Son", "Male")
p2 = Person("Keanue", "Kim", "Male")
p3 = Person("Halona", "Yun", "Female")
p1.info()
p2.info()
p3.info()
```

```python
class Student(Person):
  def __init__(self, firstname, lastname, gender, student_id, subject):
    super().__init__(firstname, lastname, gender)
    self.student_id = student_id
    self.subject = subject
```

```python
class Teacher(Person):
  def __init__(self, firstname, lastname, gender, subject, score):
    super().__init__(firstname, lastname, gender)
    self.subject = subject
    self.score = score
```

```python
X1 = [0, 0, 1, 1]
X2 = [0, 1, 0, 1]
```

```python
# 함수 and_gate(x1, x2)

def AND_gate(x1, x2):
  w1, w2, bias = 0.5, 0.5, -0.7   #and
  y = x1*w1+ x2*w2 + bias

  if y>0:
    y = 1
  else:
    y = 0
  print(f'{x1} and {x2} : {y}')
  return y
```

```python
y_and_pred = [AND_gate(x1, x2) for x1, x2 in zip(X1, X2)]
```

```python
colors = ['blue' if y == 1 else 'red' for y in y_and_pred]
plt.scatter(X1, X2, color = colors, label = 'and')
plt.xlabel(X1)
plt.ylabel(X2)
plt.legend()
```

```python
# 함수 or_gate(x1, x2)

def OR_gate(x1, x2):
  w1, w2, bias = 0.5, 0.5, -0.2  #or
  y = x1*w1+ x2*w2 + bias

  if y>0:
    y = 1
  else:
    y = 0
  print(f'{x1} or {x2} : {y}')
  return y
```

```python
y_or_pred = [OR_gate(x1, x2) for x1, x2 in zip(X1, X2)]
```

```python
colors = ['green' if y == 1 else 'yellow' for y in y_or_pred]
plt.scatter(X1, X2, color = colors, label = 'or')
plt.xlabel(X1)
plt.ylabel(X2)
plt.legend()
```

```python
# NAND_gate(x1, x2)

X1 = [0, 0, 1, 1]
X2 = [0, 1, 0, 1]

def NAND_gate(x1, x2):
  w1, w2, bias = -0.5, -0.5, 0.7    # nand
  y = x1*w1+ x2*w2 + bias

  if y>0:
    y = 1
  else:
    y = 0
  print(f'{x1} nand {x2} : {y}')
  return y

y_nand_pred = [NAND_gate(x1, x2) for x1, x2 in zip(X1, X2)]

colors = ['blue' if y == 1 else 'red' for y in y_nand_pred]
plt.scatter(X1, X2, color = colors, label = 'nand')
plt.xlabel(X1)
plt.ylabel(X2)
plt.legend()
```

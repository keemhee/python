```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#step function  ->  무시하는 값이 너무 커지면 결과값이 안 좋아져서 잘 쓰지 않음
def step_function(x):
  if x > 0:
    return 1
  else:
    return 0

X = np.arange(-3, 3, 0.1)
y = [step_function(x) for x in X]

plt.plot(X, y, label = 'step function')
plt.legend()
```

```python
# relu  ->  hidden layer 에서 주로 사용됨. 잘 모르겠으면 일단 relu로
def relu(x):
  if x > 0:
    return x
  else:
    return 0

X = np.arange(-3, 3, 0.1)
y_relu = [relu(x) for x in X]

plt.plot(X, y_relu, label = 'relu')
plt.legend()
```

```python
#sigmoid  ->  for output layer(아웃풋 결정할 때 사용). 모든 숫자가 0과 1 사이로 수렴. binary 설정할 때. 0.5를 기점으로 0일 확률, 1일 확률

def sigmoid(x):
  return 1/(1+np.exp(-x))

X = np.arange(-10, 10, 0.2)
y_sig = [sigmoid(x) for x in X]

plt.plot(X, y_sig, label = 'sigmoid')
plt.axhline(y = 0.5, color = 'red', linestyle = '--')
plt.legend()
```

```python
#softmax  ->  multi classificatio(다중분류)에서 사용

def softmax(x):
  e_x = np.exp(x - np.max(x))
  return e_x/e_x.sum(axis = 0)

X = np.arange(-10, 10, 0.2)
y_soft = [softmax(x) for x in X]

plt.plot(X, y_soft, label = 'softmax')
plt.legend()
```

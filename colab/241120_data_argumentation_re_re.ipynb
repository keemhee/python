```python
import cv2
from google.colab.patches import cv2_imshow
import tensorflow as tf
from tensorflow.keras.layers import RandomRotation, RandomZoom, RandomBrightness, RandomFlip, Resizing
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
data_argumentation = Sequential([
    RandomRotation(0.2, input_shape = (200, 200, 3)),
    RandomFlip('vertical'),#'horizontal
    RandomBrightness(0.8),
    RandomZoom(1),
    Resizing(200,200)
])
```

```python
base_dir = '/content/simple_images'
X_data = []
y_data = []
for i, class_name in enumerate(class_names):
    folder_dir = os.path.join(base_dir, class_name)

    for file_name in os.listdir(folder_dir):
      img_file = os.path.join(folder_dir, file_name)
      #print(img_file)    어느 파일에서 문제가 생긴건지 확인. 햄스터11이 문제여서 맨 위에서 지워줌
      img = cv2.imread(img_file)
      img_resized = cv2.resize(img, (200, 200))
      img_rgb = cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB)
      X_data.append(img_rgb)
      y_data.append(i)

X_data = np.array(X_data)
y_data = np.array(y_data)
print(X_data.shape)
print(y_data)
print(y_data.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dense
from tensorflow.keras.models import Sequential

model = Sequential([
    Conv2D(64, kernel_size =(5, 5), padding = 'same', strides  = (2, 2), input_shape = (200, 200, 3)), #필터 64개, 커널사이즈5x5
    MaxPool2D((2, 2)),
    Conv2D(32, kernel_size = (5, 5), padding = 'valid'),
    MaxPool2D((2, 2)),
    Flatten(),
    Dense(64, activation = 'relu'),
    Dense(32, activation = 'relu'),
    Dense(10, activation = 'softmax')
])

model.summary()

model.compile(optimizer = 'adam',
              loss = 'categorical_crossentropy',
              metrics = ['accuracy'])
history = model.fit(X_train, y_train, epochs = 10)
```

```python
def init_network():
    network = {}
    network['W1'] = np.random.randn(6).reshape((2, 3))
    network['B1'] = np.random.rand(3)
    network['W2'] = np.random.randn(6).reshape((3, 2))
    network['B2'] = np.random.rand(2).reshape((1, 2))
    network['W3'] = np.random.randn(4).reshape((2, 2))
    network['B3'] = np.random.rand(2).reshape((1, 2))
    return network

def relu(x):
    return np.maximum(0, x)

def softmax(x):
    e_x = np.exp(x - np.max(x, axis=1, keepdims=True))
    return e_x / e_x.sum(axis=1, keepdims=True)

def forward(network, x):
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    B1, B2, B3 = network['B1'], network['B2'], network['B3']

    # 입력을 2D로 변환
    x = x.reshape(1, -1)

    Z1 = np.dot(x, W1) + B1
    A1 = relu(Z1)
    Z2 = np.dot(A1, W2) + B2
    A2 = relu(Z2)
    Z3 = np.dot(A2, W3) + B3
    y = softmax(Z3)
    return y

network = init_network()
x = X_data
y_pred = forward(network, x)
print(y_pred)
```

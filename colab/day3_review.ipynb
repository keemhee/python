```python
left_count = {1: 2, 2: 3, 3: 5}
right_count = {1: 1, 2: 5, 3: 2}

for size in left_count:
  left_value = left_count[size]
  right_value = right_count[size]
  print(f'{size}의 최소 갯수: {min(left_value, right_value)}')
  # print(f'left {size} : {left_count[size]}')
  # print(f'right {size} : {right_count[size]}')
```

```python
def gloves_coupler():
  left_count = {1: 2, 2: 3, 3: 5}
  right_count = {1: 1, 2: 5, 3: 2}

  pairs = 0
  for size in left_count:
    left_value = left_count[size]
    right_value = right_count[size]
    count = min(left_value, right_value)
    pairs += count

  return pairs

answer = gloves_coupler()
print(answer)
```

 2) 학생 점수 관리 프로그램

```python
#성적관리 딕셔너리로 만들기
# 90이상A, 80이상B, 70이상C, 70미만F

grades_dic={}
my_students = ["흥민", "석구", "우성", "명신"]
grades = [80, 75, 93, 15]

for student, grade in zip(my_students, grades):
  if 90 <= grade <= 100:
    grades_dic[] += 1

  elif 80<= grade <90:
    grades_dic["B"] += 1

  elif 70<= grade <80:
    grades_dic["C"] += 1

  else:
    grades_dic["F"] += 1

print(grades_dic)

a = grades_dic.keys()
b = grades_dic.values()

plt.bar(a,b)
plt.bar(my_students,grades)
```

```python
my_students = ["흥민", "석구", "우성", "명신"]
grades = [80, 75, 93, 15]
student_grade={}

for student, grade in zip(my_students, grades):

  if grade >= 90:
    student_grade[student] = "A"
  elif grade >= 80:
    student_grade[student] = "B"
  elif grade >= 70:
    student_grade[student] = "C"
  else:
    student_grade[student] = "F"

print(student_grade)
```

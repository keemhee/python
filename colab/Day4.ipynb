# while vs for loop

```python
# 1에서 10까지의 숫자 출력하기
i = 1
while i <= 10:
  print(i)
  i += 1
```

```python
#2 짝수 출력하기
i = 1
while i <= 10:
  if i % 2 == 0:
    print(i)
  i += 1
```

```python
#3-1 1에서 100까지의 합 구하기
num = 1
sum = 0
while num <= 100:
  sum += num
  print(f"{num} : {sum}")
  num += 1
```

```python
#3-2 1에서부터 100까지 곱하기
num = 1
products = 1
while num <= 100:
  products *= num
  print(f"{num} : {products}")
  num += 1
```

```python
#4 구구단
i = 1
while i <= 9:
  j = 1
  while j <= 9:
    print(f"{i} * {j} = {i*j}")
    j+=1
  i+=1
```

```python
def union_jack():
  img = np.zeros((300, 600, 3), dtype = 'uint8')
  img[:,:] = [0, 36, 125]

  # 대각선 흰색
  for i in range(300):
    s = i*2-50; e = i*2+50;
    if s < 0:
      s = 0
    if e > 599:
      e = 599
    img[i,s:e] = [255, 255, 255]
    img[-1-i, s:e] = [255, 255, 255]

  #대각선 빨간색
  for i in range(150):
    s = i*2-30
    e = i*2+30
    if s < 0:
      s = 0
    if e> 599:
      e = 599
    img[i, s:i*2] = [207, 20, 43]
    img[-1-i, -i*2:-1-s] = [207, 20, 43]
    img[-1-i, i*2:e] = [207, 20, 43]
    img[i, -1-e:-i*2] = [207, 20, 43]

  #가로
  img[100:201, :] = [255, 255, 255]
  img[120:181, :] = [207, 20, 43]

  #세로
  img[:101, 250:351] = [255, 255, 255]
  img[201:, 250:351] = [255, 255, 255]
  img[:, 270:331] = [207, 20, 43]
```

```python
#broadcasting

import numpy as np
na5 = np.arange(24).reshape((2, 2, 2, 3))
na6 = np.arange(12).reshape((2, 2, 3))
print(na5)
print(na6)
print(na5+na6)
```

```python
# 2x2 matrix multiplication
#a @ b = a.dot(b)

na7 = np.arange(4).reshape((2,2))
na8 = np.arange(4).reshape((2,2)) * 2
print(na7)
print(na8)

print(na7 @ na8)
print(na7.dot(na8))

na10 = np.arange(48).reshape(6,8)
na11 = np.arange(16).reshape(2,8).T

print(na10 @ na11)
```

```python
#norm()
point1 = np.array([1,2])
print(point1)

d1 = np.linalg.norm(point1)
print(d1)

point2 = np.array([2,1])
point3 = point1 + point2
print(point3)

d2 = np.linalg.norm(point3)
print(d2)
```

# roll에서 shift = 1, -1
#          axis = 0, 1   의 의미

#np.roll 함수는 배열을 주어진 axis를 따라 주어진 shift만큼 이동시킴
#axis는 이동할 축, shift는 이동할 위치의 수를 나타냄
#shift = 1 이면 배열이 해당 축의 양의 방향으로 한 칸 이동
#shift = -1 이면 배열이 해당 축의 음의 방향으로 한 칸 이동
#axis = 0 이면 배열이 행 방향으로 이동

```python
#image show

import cv2
from google.colab.patches import cv2_imshow


file = '/content/cat2.jpg'
img = cv2.imread(file)
cv2_imshow(img)
print(img)

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
print(img_gray)
```

```python
#image resize

print(img.shape)

h = img.shape[0]
print(h)

aspect_r = 350 /h
print(aspect_r)

w = img.shape[1]
print(w)

img_l = cv2.resize(img, ( int(w * aspect_r), 350))
cv2_imshow(img_l)
```

```python
#image convert bgr -> rgb

# cv2 이미지는 bgr 형태
# bgr -> rgb

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2_imshow(img_rgb)
```

```python
# draw line

img_copy = img.copy()
cv2_imshow(img_copy)

h = img.shape[0]
w = img.shape[1]
cv2.line(img_copy, (0,0), (w, h), (0, 255, 0), 3, cv2.LINE_AA)
cv2_imshow(img_copy)
```

```python
# draw circle

img_copy2 = img.copy()
cv2.circle(img_copy2,
        (int(w/2), int(h/2)),
        20, (0, 0, 255), -1)
cv2_imshow(img_copy2)
```

```python
#draw a rectangle

img_copy3 = img.copy()
cv2.rectangle(img_copy3, (int(w/5), int(h/5)), (int(w/5)*4 , int(h/5)*4),
 (255, 255, 0), 5)
cv2_imshow(img_copy3)
```

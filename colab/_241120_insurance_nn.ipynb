````python
import pandas as pd
````

````python
df = pd.read_csv('/content/insurance_data.csv')
df  #data frame

#뭐가 2개ata = pd.read_csv('/content/insurance_data.csv')

#2개 인풋
#라벨 결과값 y값을 bought_insurance

#==> x 데이터와 y 데이터 나누기
````

````python
#x 데이터와 y 데이터 나누기

X_data = df[['age', 'affordibility']]
X_data.head()

y_data = df['bought_insurance']
y_data.head()
````

````python
#TENSORFLOW이용해 split.
#x데이터랑 y데이터 뭉쳐있어서 train데이터랑 test데이터 나뉘지 않음

from sklearn.model_selection import train_test_split

#X_train, X_test, y_train, y_test = train_test_split(X_data, y_data)
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size = 0.2, random_state = 42)  #8:2로 나눔.. random state하면 랜덤으로 뽑고 뽑힌 건 고정?
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
````

````python
print(X_train[:5])    #random state해서 몇 번 눌러도 같은 숫자 나옴.
                      #데이터가 들어올 때 순차적으로 들어오는 경우가 많아서 랜덤으로 뽑음.
                      #age편차가 크니 age scaling
````

````python
#age편차가 크니 age scaling


X_train['age'] = X_train['age'] / 100.0
print(X_train[:5])
X_test['age'] = X_test['age'] / 100.0
print(X_test[:5])
````

````python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
````

````python
#network구조 만들기

model = Sequential([
    Dense(1, activation = 'sigmoid', input_shape = (2, ),
          kernel_initializer = 'ones', #손으로 계산해보기위해 initial 머시기 넣어줌. 첫번째 w는 다 1
          bias_initializer = 'zeros')  #bias를 다 0으로 두겠다.
])
model.summary()
````

````python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
model.fit(X_train, y_train, epochs = 100)
````

````python
#model만들면 w주고 bias기억함. 제일 마지막의 w와 모델을 기억 -> coefficient. intercept
coef_, intercept_ = model.get_weights()
print(coef_)      #제일 마지막 w1 여기선 1.0021408, w2 0.9044141
print(intercept_) #제일 마지막 bias -0.09608371
                  #얘네 가지고 있으면 y predict 가능해짐
````

````python
def sigmoid(x):
  return 1 / (1+ np.exp(-x))

#model.predict()
def predict(age, aff):
  z = age/100.0 * coef_[0] + aff * coef_[1] + intercept_
  a = sigmoid(z)
  return a

print(predict(46, 1)) #46살이고 affordability가 1인 사람이 insurance 구매 여부 예측.
````

````python
# 0이면 + epsilon -> 최소
# 1이면 - epsilon -> 최대
# epsilon = 1e-15
#y_pred_new = np.clip(y_pred, epsilon, (1-epsilon)) 0이나 1로 가지않게 그 사이 범위로 지정
import numpy as np

def log_loss(y_true, y_pred):
    epsilon = 1e-15
    y_pred_new = np.clip(y_pred, epsilon, (1-epsilon))
    loss = -np.mean((y_true * np.log(y_pred_new)) + (1-y_true) * np.log(1-y_pred_new))
    return loss
````

````python
def gradient_descent(age, aff, y_true):
  w1 = w2 = 1
  bias = 0
  rate = 0.5
  epochs = 30
  n = len(age)

  #model.fit하는 과정
  for i in range(epochs):
    # z
    z = age * w1 + aff * w2 + bias
    # a
    y_pred = sigmoid(z)
    # loss(binary_crossentropy)
    loss = log_loss(y_true, y_pred)
    # 뒤에서부터 back propagation
    # 새로운 w1, w2, bias 찾아서 다시 for문이 새로운 w1,w2,bias로 계산되게
    # 역행렬 구하는 방법 => transpose()
    # age 역행렬
    dw1 = np.dot(np.transpose(age), (y_pred - y_true)) * (1/n)
    dw2 = np.dot(np.transpose(aff), (y_pred - y_true)) * (1/n)
    db = np.mean(y_pred - y_true)
    # w1 = w1 - something(rate * dw1)
    w1 = w1 - rate * dw1
    w2 = w2 - rate * dw2
    bias = bias - rate * db

    # 잘 됐는지 확인용
    print(f'epoch : {i}, w1 : {w1}, w2 : {w2}, bias : {bias}, loss : {loss}')

gradient_descent(X_train['age'], X_train['affordibility'], y_train)
````

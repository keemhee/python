```python
answer = 0
odd_list = []
even_list = []

# Populate odd and even lists
for i in [1, 2, 3, 4, 5, 6]:
    if i % 2 == 1:
        odd_list.append(i)
    else:
        even_list.append(i)

# Process odd list
for k in odd_list:
    # Multiply k by the power of 10 depending on its position in odd_list
    answer += k * (10 ** (len(odd_list) - odd_list.index(k) - 1))

# Process even list
for j in even_list:
    # Multiply j by the power of 10 depending on its position in even_list
    answer += j * (10 ** (len(even_list) - even_list.index(j) - 1))

print(answer)
```

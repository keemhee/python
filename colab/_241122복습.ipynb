# XOR GATE

```python
def step_function(x):
  if x <= 0:
    return 0
  else:
    return 1
```

```python
def AND_gate(x1, x2):
  w1 = 0.5
  w2 = 0.5
  b = -0.7
  a = x1 * w1 + x2 * w2 + b
  return step_function(a)

def OR_gate(x1, x2):
  w1 = 0.5
  w2 = 0.5
  b = -0.2
  a = x1 * w1 + x2 * w2 + b
  return step_function(a)

def NAND_gate(x1, x2):
  w1 = -0.5
  w2 = -0.5
  b = 0.7
  a = x1 * w1 + x2 * w2 + b
  return step_function(a)

def XOR_gate(x1, x2):
  A = NAND_gate(x1, x2)
  B = OR_gate(x1, x2)
  C = AND_gate(A, B)
  return C
```

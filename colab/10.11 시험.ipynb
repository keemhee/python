주제 : 주어진 이미지에서 특정 색상 영역(hsv)을 검출하고 해당하는 영역에 사각형을 그리기
- 색상 모델 변환(hsv)
- 트랙바를 이용하여 색상 영역 추출
- 색상 범위를 마스크 생성하여 이미지에 해당하는 색상만 남도록 필터링
- 윤곽선 검출, 사각형으로 결과 표시

```python
import cv2
import numpy as np

hsv = None
def get_hsv_value(event, x, y, flags, param):
    global hsv
    if event == cv2.EVENT_LBUTTONDOWN:  # 왼쪽 버튼 클릭
        bgr_value = frame[y, x]
        hsv_value = cv2.cvtColor(np.uint8([[bgr_value]]), cv2.COLOR_BGR2HSV)
        hsv = hsv_value[0][0]
        print(f"클릭한 위치: ({x}, {y}) - BGR: {bgr_value} - HSV: {hsv_value[0][0]}")

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()

    img_hsv =  cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    if hsv is not None:
        low_hsv = np.array([hsv[0]-10, hsv[1]-10, hsv[2]-10])
        high_hsv= np.array([hsv[0]+10, hsv[1]+10, hsv[2]+10])
        mask = cv2.inRange(img_hsv, low_hsv, high_hsv)
        img_mask = cv2.bitwise_and(frame, frame, mask = mask)
        kernel = np.ones((5,5), dtype='uint8')
        img_dilate = cv2.dilate(mask, kernel = kernel, iterations=3)
        img_erod = cv2.erode(img_dilate, kernel= kernel, iterations=3)
        img_canny = cv2.Canny(img_erod, 100, 200)


        contour, _ = cv2.findContours(img_canny, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
        c = max(contour, key = cv2.contourArea)
        img_contour = cv2.drawContours(frame, c, -1, (255, 255, 0), 3)
        x, y, w, h = cv2.boundingRect(c)
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 0, 255), 5)

        cv2.imshow('Mask', mask)
        cv2.imshow("img_dilate", img_dilate)
        cv2.imshow("img_erode", img_erod)
        cv2.imshow("img_canny", img_canny)
        cv2.imshow('img_contour', img_contour)

    if cv2.waitKey(10) & 0xff == ord('q'):
        break

    cv2.imshow("Frame", frame)
    cv2.setMouseCallback("Frame", get_hsv_value)


cap.release()
cv2.destroyAllWindows()
```

2. 숫자 맞추기 게임.
1부터 100 사이의 랜덤 숫자를 생성하고 사용자가 그 숫자를 맞추는 게임 만들기.
사용자가 입력한 숫자가 랜덤숫자보다 크거나 작은 경우 적절한 메시지를 출력.
사용자가 숫자를 맞추면 게임 종료

```python
import random

random_number = random.randint(1, 100)
attempts = 0

print("1부터 100 사이의 숫자를 맞춰보세요!")

while True:
    guess = int(input("숫자를 입력하세요: "))
    attempts += 1
    if guess < random_number:
        print("더 큰 숫자입니다.")
    elif guess > random_number:
        print("더 작은 숫자입니다.")
    else:
        print(f"축하합니다! {attempts}번 만에 맞추셨습니다.")
        break
```

```python
import random

Lotto = random.randint(0, 99)

pick = int(input("복권 번호를 입력하시오(0에서 99사이):"))
print("당첨번호는", Lotto, "입니다.")

digit1 = Lotto // 10
digit2 = Lotto % 10

pickNum1 = pick // 10
pickNum2 = pick % 10

if pick == Lotto:
    print("상금은 100만원입니다.")
elif (digit1 == pickNum1) or (digit1 == pickNum2) or (digit2 == pickNum1) or (digit2 == pickNum2) :
    print("상금은 50만원입니다.")
else:
    print("상금은 없습니다.")
```

4. 다음 코드를 grades()함수로 만들어 1번 출력.
my_students = ['흥민','석구','우성','명신]
grades = [80, 75, 93, 15]



```python
def report():
  my_students = ["흥민", "석구", "우성", "명신"]
  grades = [80, 75, 93, 15]

  grades1 = {}

  for s, g in zip(my_students, grades):
    if g >= 90:
      grades1[s] = "A"
    elif g >= 80:
      grades1[s] = "B"
    elif g >= 70:
      grades1[s] = "C"
    else:
      grades1[s] = "F"

  for key in grades1:
    print(f"{key} : {grades1[key]}")

report()
```

6. 중복횟수 출력.
1~9까지의 숫자가 무작위로 30개 나열된 리스트에서 중복횟수가 가장 많은 수와 그 수의 중복횟수 출력

```python
import random

random_list = [random.randint(1, 9) for _ in range(30)]

frequency_dict = {}

for num in random_list:
    if num in frequency_dict:
        frequency_dict[num] += 1
    else:
        frequency_dict[num] = 1

most_common_number = None
max_frequency = 0

for num in frequency_dict:
    freq = frequency_dict[num]

    if freq > max_frequency:
        most_common_number = num
        max_frequency = freq


print("리스트:", random_list)
print(f"가장 많이 중복된 숫자: {most_common_number}, 중복 횟수: {max_frequency}")
```

8. 소수판별 함수.
주어진 숫자가 소수인지 아닌지 판별하는 함수 is_prime()작성하고 예시처럼 출력.
(예시: True(7은 소수입니다.))

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

num = 7
result = is_prime(num)
print(result)  # 결과: True (7은 소수입니다.)

num = 9
result = is_prime(num)
print(result)
```

9. 1~10까지 정수 중 짝수만 출력해주는 함수 print_even, 홀수만 출력해주는 함수 print_odd를 생성하여 2개의 thread를 생성하여 동시에 실행하도록하고, thread의 작업이 끝날 때까지 대기했다가 "thread 작업 종료" 메시지 출력

```python
import threading

#짝수 출력 함수
def print_even():
  for i in range(1, 101):
    if i % 2 == 0:
      print(i)

#홀수 출력 함수
def print_odd():
  for i in range(1, 101):
    if i % 2 == 1:
      print(i)

#스레드 생성
thread1 = threading.Thread(target=print_even)
thread2 = threading.Thread(target=print_odd)

#스레드 시작
thread1.start()
thread2.start()

#스레드 종료까지 대기
thread1.join()
thread2.join()

print("Thread 작업종료")
```

12. PyQt5를 이용해 push button 2개를 만들고 누르면 버튼의 색이 변경되고 한 번 더 누르면 다시 원래의 색으로 돌아오는 toggle 버튼 만들기

```python
from PyQt5.QtWidgets import QApplication, QPushButton, QMainWindow

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.button_is_checked = False

        self.setWindowTitle("Button01")

        button = QPushButton("ON")
        button.setCheckable = True
        button.setObjectName("evilButton")
        button.clicked.connect(self.button_is_clicked)
        button.setText

        self.setCentralWidget(button)
        self.show()

    def button_is_clicked(self):
        self.button_is_checked = not self.button_is_checked       #앞의 상태와 뒤의 상태가 반대가 됨. 한 번 누르면 fasle->true 또 누르면 true->false
        if self.button_is_checked:
            self.setStyleSheet("background-color: yellow")
            self.button.setText("OFF")                            #켜졌으니 off만들기 위해선 눌러라라는 의미?
            print("LED ON")
        else:
            self.setStyleSheet("background-color: green")          #초기상태. 투명?하양?
            self.button.setText("ON")
            print("LED OFF")


app = QApplication([])
window = MainWindow()
app.exec_()
```

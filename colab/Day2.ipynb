
```
# 왜 튜플과 리스트와 세트를 오갈까
어떤 데이터는 못 바꾸게 고정시켜야 함. 예를 들어 좌표값(위도, 경도)
실수로라도 건드리면 안 되는 값은 튜플로 고정을 시켜서 다른 사람들이 바꿀 수 없게
혹시라도 고치지 못하게
중복을 확인해서 중복없게 하는 데이터세트도 필요하고
다른 사람들이 변경할 수 없게 고정해야하는 데이터세트도 필요함.
읽는 데이터로 사용할 때
```

````python
def calculator(num1, num2, operator):

  if operator == "*":
    print(num1 * num2)
  elif operator == "+":
    print(num1 + num2)
  elif operator == "-":
    print(num1 - num2)
  elif operator == "/":
    print(num1 / num2)

calculator(10, 8, "*")
````

````python
def calculator(num1, num2, operator):
    if operator  == "*" :
	    result = num1 * num2
    if operator == '/' :
      result = num1 / num2
    return result                      #result 밖으로 던져주기

my_result = calculator(10, 8, "/")     #result 받기
print(my_result)
````

```python
int button = 2;
int pins[] = {11, 10, 9, 8};
bool flag = false;
int pin = 0; // 현재 핀 인덱스

void setup() {
  pinMode(button, INPUT_PULLUP);

  for (int i = 0; i < 4; i++) {
    pinMode(pins[i], OUTPUT);
  }
  attachInterrupt(digitalPinToInterrupt(button), onnoff, CHANGE);
}

void loop() {
  if (flag) {
    digitalWrite(pins[pin], LOW);
    pin = (pin + 1) % 4;
    digitalWrite(pins[pin], HIGH);
    delay(1000); //
  }
}

void onnoff() {
  flag = !flag;
  pin = 0;
  for (int i = 0; i < 4; i++) {
    digitalWrite(pins[i], LOW);
  }
  digitalWrite(pins[pin], HIGH);
}
```

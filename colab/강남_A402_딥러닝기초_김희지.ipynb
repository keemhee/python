```python
import numpy as np

def relu(x):
  return np.maximum(0, x)

def sigmoid(x):
  return 1/(1+np.exp(-x))

X1 = np.array([1, 0.5])
W1 = np.random.randn(6).reshape((2, 3))
B1 = np.random.rand(3)
Z1 = np.dot(X1, W1) + B1
A1 = relu(Z1)
W2 = np.random.randn(12).reshape((3, 4))
B2 = np.random.rand(4)
Z2 = np.dot(A1, W2) + B2
A2 = relu(Z2)
W3 = np.random.randn(8).reshape((4, 2))
B3 = np.random.rand(2)
Z3 = np.dot(A2, W3) + B3
A3 = relu(Z3)
W4 = np.random.randn(2).reshape((2, 1))
B4 = np.random.rand(1)
Z4 = np.dot(A3, W4) + B4
y_pred = sigmoid(Z4)
print(y_pred)
```

```python
# 어디에.append(집어넣을 아이)**********가장 마지막에 한 개씩 추가된다.어펜드하고 나선 프린트 꼭 해줘야 한다.
numbers_add_2 = []
for i in range(6) :
  numbers_add_2.append(numbers[i] + 2)
print(numbers_add_2)

numbers_add_3 = numbers_add_2
print(numbers_add_3[2:5])

numbers_add_4 = numbers_add_2[2:5]
print(numbers_add_4)

numbers_add_5 = []
for i in range(3) :
  numbers_add_5.append(numbers[i] + 4)
print(numbers_add_5)

for num in numbers[1:4]:
  print(num + 2)
```

```python
nums = []
for i in range(10):
  nums.append(i+1)

print(nums)

print(nums[2:8])
print(nums[-1])
print(nums[-3])
print(nums[-3:])

odd_nums = nums[::2]
print(odd_nums)
even_nums = nums[1::2]
print(even_nums)

first_5_nums = nums[ : 5]
reversed_first_5_nums = nums[:5][::-1]
print(reversed_first_5_nums)

first_5_nums2 = nums[ : 5]
reversed_first_5_nums2 = first_5_nums2[::-1]
print(reversed_first_5_nums2)

print(nums[4::-1])

nums.reverse()
print(nums[5:])
```

```python
menu1 = ["칼국수", "랍스터", "피자", "쌀국수"]
menu2 = ['샌드위치', '짜장면', '김치찌개', '곱창']
menu3 = ['아메리카노', '김치랑밥', '김치찌개', '삼각김밥']
menu4 = ['유부초밥', '김밥', '국밥', '짬뽕']

menu = []
menu.extend(menu1)
menu.extend(menu2)
menu.extend(menu3)
menu.extend(menu4)
print(menu)
print(len(menu))

#겹치는 거 빼기  // set은 중복을 없애줌. 중복 없이 나열. 순서 상관x. 부를 수도 없어. 하나하나 부르는 의미는 없어서 인덱싱 안 됨 -> 다시 리스트로 만들어줌
menu2 = set(menu)
print(menu2)
print(len(menu2))

menu2_list = list(menu2)
```

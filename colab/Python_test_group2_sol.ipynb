#문제 1. 두 리스트의 교집합
두 개의 리스트가 주어질 때, 이 두 리스트의 교집합을 반환하는 함수를 작성하세요.

```python
def intersection(list1, list2):
    intersection_list = []

    for a, b in zip(list1, list2):
        if a in list2 and a not in intersection_list:
            intersection_list.append(a)
        if b in list1 and b not in intersection_list:
            intersection_list.append(b)
        return intersection_list

list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
output = intersection(list1, list2)
print(output)
```

# 문제2. 마름모 형태의 별 찍기
주어진 홀수 n에 대해 마름모 형태의 별을 출력하는 프로그램을 작성하세요. 예를 들어 n=5일 때, 다음과 같은 결과를 출력해야 합니다.




화면 캡처 2024-10-04 165014.png




```python
def print_diamond(n):
    # 위쪽 삼각형
    for i in range(n // 2 + 1):
        # 공백과 별을 한 번에 출력하는 방식
        line = ' ' * (n // 2 - i) + '*' * (2 * i + 1)
        print(line)

    # 아래쪽 삼각형
    for i in range(n // 2 - 1, -1, -1):
        line = ' ' * (n // 2 - i) + '*' * (2 * i + 1)
        print(line)

# 예제 사용
n = 5
print_diamond(n)
```

#문제4. 중첩 리스트에서 최대, 최소값 구하기
[[3, 5, 7], [1, 6, 8], [4, 9, 2]] 이렇게 리스트가 주어졌습니다.


```python
nested_list = [[3, 5, 7], [1, 6, 8], [4, 9, 2]]
# 최댓값과 최솟값을 첫 번째 요소로 초기화
max_val = nested_list[0][0]
min_val = nested_list[0][0]

# 중첩 리스트를 하나씩 순회
for inner_list in nested_list:
    # 각 내부 리스트의 최댓값과 최솟값을 계산
    max_in_list = max(inner_list)
    min_in_list = min(inner_list)

    # 최댓값, 최솟값을 전체 최댓값, 최솟값과 비교하여 갱신
    if max_in_list > max_val:
        max_val = max_in_list
    if min_in_list < min_val:
        min_val = min_in_list

print(f"최댓값: {max_val}, 최솟값: {min_val}")
```

#문제 5. 숫자 맞추기 게임
1부터 100 사이의 랜덤 숫자를 생성하고 사용자가 그 숫자를 맞추는 게임을 만드세요. 사용자가 입력한 숫자가 랜덤 숫자보다 크거나 작은 경우 적절한 메시지를 출력합니다. 사용자가 숫자를 맞추면 게임이 종료됩니다.


```python
import random

random_number = random.randint(1, 100)
attempts = 0

print("1부터 100 사이의 숫자를 맞춰보세요!")

while True:
    guess = int(input("숫자를 입력하세요: "))
    attempts += 1
    if guess < random_number:
        print("더 큰 숫자입니다.")
    elif guess > random_number:
        print("더 작은 숫자입니다.")
    else:
        print(f"축하합니다! {attempts}번 만에 맞추셨습니다.")
        break
```

#문제7. 이미지 블러링
주어진 이미지에 가우시안 블러를 적용하고 결과 이미지를 저장하세요. 블러링 커널 크기는 5x5로 설정합니다.

```python
import cv2

def apply_gaussian_blur(image_path, output_path):
    # 이미지 불러오기
    image = cv2.imread(image_path)
    # 가우시안 블러 적용
    blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
    # 이미지 저장
    cv2.imwrite(output_path, blurred_image)
```

#문제 8. 이미지 비교
주어진 이미지를 그레이스케일로 변환한 후, 트랙바를 이용하여 이진화 임계값을 조정하면서 원본 이미지와 이진화된 이미지를 실시간으로 비교할 수 있도록 하세요. 사용자가 임계값을 조정하여 특정 객체를 강조할 수 있도록 구현해야 합니다.

요구사항
1. 사용자가 조정할 수 있는 트랙바를 만들 것.
2. 원본 이미지와 이진화된 이미지를 두 개의 창에 동시에 표시할 것.
3. 'q' 키를 누르면 프로그램이 종료되도록 할 것.



```python
import cv2

# 이미지 파일 경로
file = 'cat.jpg'

# 이미지를 그레이스케일로 읽기
img_gray = cv2.imread(file, cv2.IMREAD_GRAYSCALE)

def do_nothing(value):
    pass

# 트랙바가 적용될 창 생성
cv2.namedWindow("Trackbar")
# 트랙바 생성: 이름, 창 이름, 초기값, 최대값, 콜백 함수
cv2.createTrackbar("Threshold", "Trackbar", 100, 255, do_nothing)

while True:
    # 트랙바에서 현재 임계값 가져오기
    thresh_value = cv2.getTrackbarPos("Threshold", "Trackbar")
    # 임계값을 이용해 이미지를 이진화
    _, img_thresh = cv2.threshold(img_gray, thresh_value, 255, cv2.THRESH_BINARY)

    # 원본 이미지와 이진화된 이미지 표시
    cv2.imshow("Cat", img_gray)
    cv2.imshow("Threshold", img_thresh)

    # 'q' 키를 누르면 종료
    if cv2.waitKey(5) & 0xFF == ord('q'):
        break

# 모든 윈도우 닫기
cv2.destroyAllWindows()
```

#문제9. 중복횟수 출력
1~9까지의 숫자가 무작위로 30개 나열된 리스트에서 중복횟수가 가장 많은 수와 그 수의 중복횟수 출력


```python
import random

random_list = [random.randint(1, 9) for _ in range(30)]

frequency_dict = {}

for num in random_list:
    if num in frequency_dict:
        frequency_dict[num] += 1
    else:
        frequency_dict[num] = 1

most_common_number = None
max_frequency = 0

for num in frequency_dict:
    freq = frequency_dict[num]

    if freq > max_frequency:
        most_common_number = num
        max_frequency = freq


print("리스트:", random_list)
print(f"가장 많이 중복된 숫자: {most_common_number}, 중복 횟수: {max_frequency}")
```

#문제 10. 1부터 100까지 숫자들을 자리가 같은 수끼리 분류하세요.

```python
grouped_numbers = {}

for num in range(1, 101):
    second_digit = (num // 10) % 10

    if second_digit not in grouped_numbers:
        grouped_numbers[second_digit] = []

    grouped_numbers[second_digit].append(num)

for second_digit in grouped_numbers:
    numbers = grouped_numbers[second_digit]
    print(f"둘째자리가 {second_digit}인 숫자들: {numbers}")
```

#문제 11. 기한만료 여부 판단
 한달을 30일로 계산, 년월일을 YYYY.MM.DD로 입력받았을때 2024.10.04와 일수의 차이가 6개월 이상일 때 "기한만료", 미만일 때 "유효"를 출력하는 함수


```python
# 날짜를 일 단위로 변환하는 함수 (YYYY.MM.DD)
def convert_to_days(date):
    year, month, day = map(int, date.split('.'))
    return year * 12 * 30 + month * 30 + day

# 기한 만료 여부를 체크하는 함수
def check_expiry(input_date):
    # 기준 날짜 (2024.10.04)를 일 단위로 변환
    base_date = "2024.10.04"
    base_in_days = convert_to_days(base_date)

    # 입력받은 날짜를 일 단위로 변환
    input_in_days = convert_to_days(input_date)

    # 일수 차이 계산
    day_difference = base_in_days - input_in_days

    # 6개월 = 168일 차이
    if day_difference >= 180:
        print("기한만료")
    else:
        print("유효")

check_expiry("2024.10.11")
```

#문제14. 소수판별 함수
주어진 숫자가 소수인지 아닌지를 판별하는 함수 is_prime을 작성하고 예시처럼 출력하세요.
예시: True (7은 소수입니다.)


```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

num = 7
result = is_prime(num)
print(result)  # 결과: True (7은 소수입니다.)

num = 9
result = is_prime(num)
print(result)
```

#문제16. 복권 상금 알려주기
2자리 숫자로 이루어진 복권이 있습니다. 사용자가 가지고 있는 복권 번호가 2자리 모두 일치하면 100만 원을 받습니다. 2자리 중에서 하나만 일치하면 50만 원을 받습니다. 하나도 일치하지 않으면 상금은 없습니다. 복권 당첨 번호는 난수로 생성하고 사용자의 입력에 따라서 상금이 얼마인지를 출력하는 프로그램을 작성하세요.


```python
import random

Lotto = random.randint(0, 99)

pick = int(input("복권 번호를 입력하시오(0에서 99사이):"))
print("당첨번호는", Lotto, "입니다.")

digit1 = Lotto // 10
digit2 = Lotto % 10

pickNum1 = pick // 10
pickNum2 = pick % 10

if pick == Lotto:
    print("상금은 100만원입니다.")
elif (digit1 == pickNum1) or (digit1 == pickNum2) or (digit2 == pickNum1) or (digit2 == pickNum2) :
    print("상금은 50만원입니다.")
else:
    print("상금은 없습니다.")
```
